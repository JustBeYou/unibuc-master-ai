```python
def plot_degrees(input_text):
    lines = [line.strip() for line in input_text.split('\n') if line.strip() != '']
    q = [line.replace('Cuvrve for ', '').replace(': ', '|').split('|') for line in lines]
    k = [[x[0], eval(x[1])] for x in q]

    fig, ax = plt.subplots(nrows=3, ncols=3, dpi=150, figsize=(6.4*2, 4.8*2))

    i = 0
    for row in ax:
        for col in row:
            if i >= len(k):
                break
            
            z = k[i]
            i+=1
            col.plot(z[1], label=z[0])
            col.legend()
```

```python
plot_degrees(input_text)
```
